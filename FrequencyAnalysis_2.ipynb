##Single letter Occurance

```python
from collections import defaultdict
occurance = defaultdict(int)
for i in cryptogram:
  occurance[i] +=1
occurance.pop(" ")
occurance= dict(sorted(occurance.items(), key=lambda item: item[1],reverse=True))
```

```python
from prettytable import PrettyTable as pt
table = pt()
table.add_row(occurance.keys())
table.add_row(occurance.values())
print(table)
```

```python
import matplotlib.pyplot as plt
letters = occurance.keys() 
freq = occurance.values()
plt.figure(figsize=(15,5))
plt.bar(letters,freq)
```

#**R13 Substitution method**

####To apply ROT13 to a piece of text, take every letter in the Latin alphabet and replace it by the letter 13 places further along in the alphabet, wrapping back to the beginning if necessary, preserving case: a becomes n, B becomes O, and so forth, down to Z, which becomes M. Numbers, symbols, and other characters are left unchanged.


```python

actual_letter ="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"            # In rot13 every letter is rotate 13 letters forward
cipher_letter = "NOPQRSTUVWXYZABCDEFGHIJKLMnopqrstuvwxyzabcdefghijklm"   

decipher = ''

for i in cryptogram:
  if i not in cipher_letter:
   decipher += i
  else:
    decipher += actual_letter[cipher_letter.index(i)]
    
print(cryptogram)
print("\nDecryted text is > \n")
print(decipher)
```
